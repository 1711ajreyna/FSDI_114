```python
# properties:
   # head: front of queue
    #tail: back of queue
    #__size: private variable  to store data

#Constructor: 
# Since initially a new queue is created as empty
    # initilize head and tail to None
    # Initilize __size to zero 

#Methods:
    # enqueue(value):
        # Create a new node with a value
        # If the queue is empty:
            # set front and rear to the new node since it's only one element
        # Else:
            # Set the next pointer of the current rear to the new node 
            # make it so the current node at the rear points to the new node 
        # increment size 

    # dequeue():
        # If the queue is empty:
           # make an exception indicating the queue is empty
        # get the value of the node at the head of the queue
        # move the head pointer to the next node
        # if the head becomes None:
            # set the rear to None
        # decrement __size
        # return the value

    # size():
        # return the value of __size

    # peek():
        # if thd queue is empty:
            # indicate the queue is empty
        # return the value of the node at the head of the queue

    # is_empty():
        # return True if __size is 0,  else return False

# create a Node class to represent each element in queue 
class Node:
    def __init__(self, value):
        self.value = value     # the value of the node 
        self.next = None       # reference the next node in the queue


class QueueII:
    # initilize head and tail to none, __size to zero
    #
    def __init__(self):
        self.front = None   # head to ref the first node in queue
        self.rear = None   # tail to ref the last node in queue
        self.__size = 0    # private variable to store the size of the queue

    # creates a new node with the provided value and adds it to the end of the queue. updating the tail pointer.
    def enqueue(self, value):
        new_node = Node(value)   # create a new instance of Node
        if self.front is None:    # determine if the state of the queue, if it's empty or not.  if it is set rear = front since same node
            self.front = new_node
            self.rear = new_node
        else:                          # if it's not empty
            self.rear.next = new_node  # rear node points to the new node
            self.rear = new_node       # update the nodes rear reference
        self.__size += 1
    
    # remove and return the value of the node at the front end of the queue and update the pointer
    def dequeue(self):
        if self.front is None:                # if the queue is empty raise th exception "Queue is empty"
            raise Exception("Queue is empty")
        datum = self.front.value               # gets the value of the node at the head of the queue
        self.front = self.front.next           # move the head pointer to the next node
        if self.front is None:                # If the queue becomes empty set the rear node to None as well
            self.rear = None
        self.__size -= 1                       # decrement by 1 since a node was removed
        return datum
    
    # return the current size of the queue
    def size(self):
        return self.__size
    
    # return the value of the node at the front of the queue without removing it
    def peek(self):
        if self.front is None:                # if the queue is empty
            raise Exception("Queue is empty")
        return self.front.value
    
    # check if the queue is empty
    def is_empty(self):
        return self.__size == 0 # returns True if empty and False if it's not

queue = QueueII()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.dequeue()) # should print 1
print(queue.dequeue()) # 2
print(queue.dequeue()) # 3

print(queue.dequeue()) # none
```

```python
# class SinglyLinkedList:
#  class __Node:
#       data stored in node
#       reference to the next node in the list

#       __Node(data): constructor for creating new node
#         initilize node data by setting data to data
#         initilize next node ref to None by setting next to none

#  head - ref to first node in list
#  tail - ref to last node in list

# SinglyLinkedList():
#     initilize head ref to None
#     initilize tail ref to None

# append(value):
#     new_node = create a instance of __Node with value
#     if head is None:
#         set head to new_node
#         set tail to new_node
#     else:
#         set the next ref of the current tail node to the new node
#         update tail to point to the new node

# insert(index, value): 
#     check if index is negative
#     if so raise an exception
#     if index == 0:
#        new_node = create a new instance of __Node with value
#        set the next ref of the new node to the current head 
#        update the head to point at the new node
#        if the list is empty
#              update tail to new_node 
#     else:
#          set current to head 
#          init index counter, i = 0
#          travel the index untill the previous node to the insert point
#               move current to the next node
#               increment index ounter, i += 1
#          create a new node 
#          set the next reference of the new node to the next node of the current node
#          update the next reference of the current node to the new node
#          if the new node is inserted at end of the list:
#                  update the tail to point at the new node 

# __str__():
#    init an empty string out with "[" at the beginning 
#    set current to the head node of the list
#    if current is not None:
#        append the string representation of the data of the current node to out 
#        Move the current to the next node 
#        while current is not None:
#           append ", " and the string representation of the data of the current node to out
#           move current to the next node
#     append "]" to out
#     return out


class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None  # Initialize tail reference
        
# challenge yourself:  see if you can ahave and maintatin a tail ref and how that impacts performance
    # Modify the SingleLinkedList class to maintain a tail reference, which points to the last node in the list.
    # This should improve the performance of appending operations, allowing them to run in a constant time.

    
    def append(self, value):
        new_node = self.__Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node  # Updates tail for the first node
        else:
            self.tail.next = new_node
            self.tail = new_node  # Updates tail to the new node
            
# this method has to insert a new node (with value) before the index, not after!
    def insert(self, index, value):
        if index < 0:
            raise ValueError("index must be non-negative")
        if index == 0:
            new_node = self.__Node(value)
            new_node.next = self.head
            self.head = new_node
            if self.tail is None:
                self.tail = new_node   # Updates tail if the list was empty
        else:
            current = self.head
            i = 0
            while current is not None and i < index - 1:
                current = current.next
                i += 1
            new_node = self.__Node(value)
            new_node.next = current.next
            current.next = new_node
            if new_node.next is None:
                self.tail = new_node    # Update tail if the new node is insterted at the tail
    
    def __str__(self):
        # this dunder method must return a string at all times
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out

sll = SinglyLinkedList()
sll.append(1)
sll.append(2)
sll.append(3)

print("Original list:", sll)

sll.insert(1, 6)

print("After inserting 6 at index 1:", sll)
```
