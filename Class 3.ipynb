```python
# properties:
   # head: front of queue
    #tail: back of queue
    #__size: private variable  to store data

#Constructor: 
# Since initially a new queue is created as empty
    # initilize head and tail to None
    # Initilize __size to zero 

#Methods:
    # enqueue(value):
        # Create a new node with a value
        # If the queue is empty:
            # set front and rear to the new node since it's only one element
        # Else:
            # Set the next pointer of the current rear to the new node 
            # make it so the current node at the rear points to the new node 
        # increment size 

    # dequeue():
        # If the queue is empty:
           # make an exception indicating the queue is empty
        # get the value of the node at the head of the queue
        # move the head pointer to the next node
        # if the head becomes None:
            # set the rear to None
        # decrement __size
        # return the value

    # size():
        # return the value of __size

    # peek():
        # if thd queue is empty:
            # indicate the queue is empty
        # return the value of the node at the head of the queue

    # is_empty():
        # return True if __size is 0,  else return False

# create a Node class to represent each element in queue 
class Node:
    def __init__(self, value):
        self.value = value     # the value of the node 
        self.next = None       # reference the next node in the queue


class QueueII:
    # initilize head and tail to none, __size to zero
    #
    def __init__(self):
        self.front = None   # head to ref the first node in queue
        self.rear = None   # tail to ref the last node in queue
        self.__size = 0    # private variable to store the size of the queue

    # creates a new node with the provided value and adds it to the end of the queue. updating the tail pointer.
    def enqueue(self, value):
        new_node = Node(value)   # create a new instance of Node
        if self.front is None:    # determine if the state of the queue, if it's empty or not.  if it is set rear = front since same node
            self.front = new_node
            self.rear = new_node
        else:                          # if it's not empty
            self.rear.next = new_node  # rear node points to the new node
            self.rear = new_node       # update the nodes rear reference
        self.__size += 1
    
    # remove and return the value of the node at the front end of the queue and update the pointer
    def dequeue(self):
        if self.front is None:                # if the queue is empty raise th exception "Queue is empty"
            raise Exception("Queue is empty")
        datum = self.front.value               # gets the value of the node at the head of the queue
        self.front = self.front.next           # move the head pointer to the next node
        if self.front is None:                # If the queue becomes empty set the rear node to None as well
            self.rear = None
        self.__size -= 1                       # decrement by 1 since a node was removed
        return datum
    
    # return the current size of the queue
    def size(self):
        return self.__size
    
    # return the value of the node at the front of the queue without removing it
    def peek(self):
        if self.front is None:                # if the queue is empty
            raise Exception("Queue is empty")
        return self.front.value
    
    # check if the queue is empty
    def is_empty(self):
        return self.__size == 0 # returns True if empty and False if it's not

queue = QueueII()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.dequeue()) # should print 1
print(queue.dequeue()) # 2
print(queue.dequeue()) # 3

print(queue.dequeue()) # none
```
